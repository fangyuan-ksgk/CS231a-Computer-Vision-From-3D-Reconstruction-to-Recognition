```python
from skimage import io
import skimage as sk
from matplotlib import pyplot as plt
```

```python
img = io.imread('image1.jpg',as_gray=True)
img = sk.img_as_float(img)
# singular value decomposition
u,s,vh = np.linalg.svd(img,full_matrices=True)
# rank 1 approximation of image
img_r1 = s[0] * (u[:,:1] @ vh[:1])
# rank r approximation of image
r = 20
temp = u[:,:r].transpose(1,0)[:,:,np.newaxis] @ vh[:r,:][:,np.newaxis,:]
img_r = np.sum(temp,axis=0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
plt.imshow(img_r1,cmap='gray')
```
